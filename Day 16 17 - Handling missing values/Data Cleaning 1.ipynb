    <h3>   What Are Missing Values?

In a dataset, sometimes some entries are **missing**.

They may appear as:

- `NaN` (Not a Number)
- `None`
- Empty cells

#### Why do missing values happen?

Real-world reasons:

- Person did not answer a question in a form
- Sensor was offline and did not record data
- Data entry mistake
- Different sources joined together, some fields not available

#### Why do we care?

Many algorithms and functions:

- **Cannot handle** missing values directly
- Will give errors or wrong results if we ignore them

So we must **detect** and **handle** missing values properly.

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3>   Common Strategies to Handle Missing Values

There is **no single correct way**. It depends on your data and problem.

#### Option 1 – Drop Missing Values

- Remove rows that have missing values
- Or remove columns with too many missing values

Pros:
- Simple
- Safe if only few rows/columns are removed

Cons:
- You may lose **a lot of data**
- Can create bias if removed rows are not random


#### Option 2 – Impute (Fill In) Missing Values

- Replace missing values with some reasonable value:

For numeric columns:
- **Mean** (average)
- **Median** (middle value, more robust to outliers)

For categorical columns:
- **Mode** (most frequent value)
- A special label, e.g., `"Unknown"`

Pros:
- Keeps most/all rows
- Simple to implement

Cons:
- We are **guessing** values
- Too much imputation can distort the data

In practice, people often **combine** both:
- Drop columns/rows with *too many* missing values
- Impute missing values in important columns

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  What Are Outliers?

**Outliers** are values that are **unusually far** from most of the other values.

Examples:

- Most people’s heights are between 140–200 cm; one value is 260 cm → suspicious  
- Most transaction amounts are 10–500 dollars; one is 50,000 dollars → may be an outlier

#### Why do outliers appear?

- Genuine rare events (e.g., very big purchase)
- Data entry errors (e.g., typing 1000 instead of 100)
- Measurement errors (faulty sensor)

#### Why do we care?

Outliers can:

- **Skew** the mean and standard deviation
- Affect models like **Linear Regression**, **K-Means**, etc.
- Mislead visualizations and summary statistics

But:

- Not all outliers are bad  
  Some contain **important information** (like fraud detection)

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   Ways to Detect Outliers (Basic)

#### 1. Summary Statistics

- Look at:
  - `min`, `max`
  - `mean`
  - `std` (standard deviation)
  - `25%`, `50%`, `75%` percentiles

If max is **very far** from 75% percentile, it may be an outlier.

When we talk about 25%, 50%, 75% in IQR and boxplots, we’re really talking about percentiles or quartiles:
- 25% → 25th percentile → Q1 (First Quartile) = upto the first 25% of data
- 50% → 50th percentile → Q2 (Median)
- 75% → 75th percentile → Q3 (Third Quartile)

#### 2. Boxplot

A **boxplot** shows:

- Median (middle)
- 25th and 75th percentiles (Q1 and Q3)
- “Whiskers” that often extend to 1.5 × IQR

**IQR (Interquartile Range)** = Q3 – Q1

A common rule:

- Values below `Q1 - 1.5 * IQR` → lower outliers
- Values above `Q3 + 1.5 * IQR` → upper outliers

This is a simple, widely used rule.

    <h4>   Let’s Use a Real Dataset: Seaborn `titanic`

We will work with the `titanic` dataset:

- Each row is a person on the Titanic
- Columns include:
  - `age`
  - `fare`
  - `sex`
  - `class`
  - `embarked`
  - `survived`
- It has **missing values** and also some **outliers** (e.g., in `fare`)


```python
# Import required libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (7, 5)
```

```python
# Load the built-in Titanic dataset

titanic = sns.load_dataset("titanic")

# Look at first few rows
titanic.head()
```

```python
# Basic info about the dataset

titanic.info()
```

##### What do we see?

- `age`, `deck`, `embark_town` and maybe other columns have **missing values**
- `fare` is a numeric column that may contain **outliers** (very high values)

We will work mainly with:

- `age` (has missing values)
- `embarked` or `embark_town` (categorical with missing values)
- `fare` (to look at outliers)


#### Detecting Missing Values

Pandas gives us easy ways to see missing values.

```python
# 6.1 Total missing values in each column

titanic.isna().sum()
```

```python
# Percentage of missing values in each column

missing_percent = titanic.isna().mean() * 100
missing_percent.round(2)
```

#### Observations

- Which columns have the most missing values?
- Are there columns with more than, say, 40–50% missing?
  - Those might be candidates for **dropping**.


<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3>  Simple Strategy: Dropping Missing Values

We will try two examples:

1. Drop all rows that have *any* missing values in important columns (like `age`)
2. Drop a whole column that is mostly missing (like `deck`)

```python
# Drop rows where 'age' is missing

titanic_age_not_null = titanic.dropna(subset=["age"])

print("Original shape:", titanic.shape)
print("After dropping rows with missing age:", titanic_age_not_null.shape)
```

```python
# Drop 'deck' column entirely (if we decide it is too incomplete)

titanic_no_deck = titanic.drop(columns=["deck"])

print("Columns before dropping:", titanic.columns.tolist())
print("Columns after dropping 'deck':", titanic_no_deck.columns.tolist())
```

#### When to drop?

- Drop rows if:
  - Only a **small percentage** of rows are affected
  - The missing values seem random

- Drop columns if:
  - Most of the column is missing
  - The column is not very important for the analysis or model


<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3> Imputing Missing Values (Filling In)

Now we will **fill** missing values instead of dropping.

We will:

1. Fill numeric column `age` with its **median**  
2. Fill categorical column `embark_town` with its **mode** (most frequent value)

```python
# Check 'age' summary before imputation

titanic["age"].describe()
```

```python
# Fill missing 'age' with median age

titanic_imputed = titanic.copy()

median_age = titanic_imputed["age"].median()
print("Median age:", median_age)

titanic_imputed["age"] = titanic_imputed["age"].fillna(median_age)

# Check if there are still missing ages
titanic_imputed["age"].isna().sum()
```

```python
# Impute 'embark_town' with its mode (most frequent value)

titanic_imputed["embark_town"].value_counts(dropna=False)
```

```python
# Fill missing 'embark_town' with mode

mode_embark_town = titanic_imputed["embark_town"].mode()[0]
print("Mode of embark_town:", mode_embark_town)

titanic_imputed["embark_town"] = titanic_imputed["embark_town"].fillna(mode_embark_town)

# Check missing values again
titanic_imputed[["age", "embark_town"]].isna().sum()
```

#### Key Point

Imputation introduces **guessed** values.  
So always remember:

- Document what you did
- Be careful when lots of values are missing
- In real ML, we often use tools like `SimpleImputer` (later in the course)


#### Detecting Outliers – Using `fare` as Example

We will now look at **outliers** in the `fare` column.

Steps:

1. Look at summary statistics
2. Plot a boxplot
3. Use the IQR rule to find outliers

```python
# Summary statistics of 'fare'

titanic["fare"].describe().round(2)
```

```python
# Boxplot of 'fare'

sns.boxplot(x=titanic["fare"])
plt.title("Boxplot of Fare")
plt.xlabel("Fare")
plt.show()
```

#### Observations

- Most fares are low
- A few fares are **very high** (right side long whisker and points)
- These high fares are outliers according to the boxplot

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3> Finding Outliers Using IQR Rule

Recall:

- Q1 = 25th percentile  
- Q3 = 75th percentile  
- IQR = Q3 – Q1

Outlier rule:

- Lower bound = Q1 – 1.5 * IQR  
- Upper bound = Q3 + 1.5 * IQR  
- Values outside this range are considered outliers (simple rule)

```python
# Compute IQR for fare

Q1 = titanic["fare"].quantile(0.25)
Q3 = titanic["fare"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", round(Q1, 2))
print("Q3:", round(Q3, 2))
print("IQR:", round(IQR, 2))
print("Lower bound:", round(lower_bound, 2))
print("Upper bound:", round(upper_bound, 2))
```

```python
# Count how many outliers we have

outliers_high = titanic[titanic["fare"] > upper_bound]
outliers_low = titanic[titanic["fare"] < lower_bound]

print("Number of high outliers (fare > upper_bound):", outliers_high.shape[0])
print("Number of low outliers (fare < lower_bound):", outliers_low.shape[0])
```

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3> Handling Outliers – Simple Approaches

1. **Remove** rows with outliers  
   - Example: keep only rows where `fare` is between lower and upper bound

2. **Cap** outliers  
   - Example: any value above upper_bound is set to upper_bound  
   - Any value below lower_bound is set to lower_bound

3. **Leave them**  
   - If they are real and important (e.g., very rich passengers)

We will show option 1 (remove) and option 2 (cap).

```python
# Remove rows with 'fare' outliers (keep only non-outliers)

titanic_no_outliers = titanic[(titanic["fare"] >= lower_bound) & (titanic["fare"] <= upper_bound)]

print("Original shape:", titanic.shape)
print("After removing outliers in fare:", titanic_no_outliers.shape)
```

```python
# Cap 'fare' values at the upper bound (Winsorizing)

titanic_capped = titanic.copy()
titanic_capped["fare_capped"] = titanic_capped["fare"].clip(lower=lower_bound, upper=upper_bound)

# Compare original vs capped with a boxplot
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.boxplot(x=titanic["fare"])
plt.title("Original Fare")

plt.subplot(1, 2, 2)
sns.boxplot(x=titanic_capped["fare_capped"])
plt.title("Capped Fare")

plt.tight_layout()
plt.show()
```

#### Observation

- After **removing** outliers → fewer rows, but data range is "cleaner"
- After **capping** outliers → we keep all rows, but **extreme values are reduced**

In real work:

- You decide based on:
  - Context (what does fare represent?)
  - Model sensitivity to outliers
  - How many outliers there are
