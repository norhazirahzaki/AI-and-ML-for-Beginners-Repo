```python
import numpy as np

x = [20,50,60,40,30,50,40,3,110]

Q1 = np.percentile(x,25)

Q2 = np.percentile(x,50)

Q3 = np.percentile(x,75)

IQR = Q3 - Q1

upper_lmt = Q3 + (1.5 * IQR)   # 80

lower_lmt = Q1 - (1.5 * IQR)   # 0

print(upper_lmt, lower_lmt)
```

```python
new_list = []

for i in x:
    if i > lower_lmt  and  i < upper_lmt:   # window = from 0   ----- >   80
        new_list.append(i)

new_list
```

```python
# list comprehensions

# from lower_lmt to upper_lmt
# i < upper  i> lower
x_clean = [ i for i in x if i < upper_lmt and i > lower_lmt ]

x_clean = [ i for i in x if lower_lmt <= i <= upper_lmt ]

x_clean
```
