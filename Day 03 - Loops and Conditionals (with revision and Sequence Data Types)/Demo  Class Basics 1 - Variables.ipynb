### Variables in python

- variables are used to store data in a program.

```python
import sys

a = 1 # integer
b = 1.44 # float
c = " " # string

d = [1,2] # list

print("int:", sys.getsizeof(a), "bytes")
print("float:", sys.getsizeof(b), "bytes")
print("str:", sys.getsizeof(c), "bytes")
print("list:", sys.getsizeof(d), "bytes")
```

```python
# store data in variables
# writing the name:
# use proper full names and not abbreviations
# use lowercase letters
# Snake_case: use underscores to separate words in variable names
# example: first_name, roll_no, address
# CamelCase for functions and classes: 
```

```python
first_name = "idrees"
roll_no = 23
address = "kashmir"

# examples of bad variable names 
#nm = "idrees"
#rn = 23
#ad = "kashmir"
```

```python
# printing

print("Hello, " + first_name) # concatenation of strings

# formmatted strings
print(f"hello, {first_name}")

# using the .format() method
print("Hello, {}. Your roll number is: {}".format(first_name, roll_no))

# using %s - string type or %d - integer type
print("Hey, %s ! your roll number is: %d "%(first_name, roll_no)) 

# printing float values of different precisions
pi = 3.141592653589793

print("The value of pi is approximately: %.2f" %pi)  # limiting to 2 decimal places
print("The value of pi is approximately: %.5f" %pi)  # limiting to 5 decimal places
print("The value of pi is approximately: %.10f" %pi)  # limiting to 10 decimal places

# using f-strings for float formatting
print(f"The value of pi is approximately: {pi:.2f}")  # limiting to 2 decimal places
print(f"The value of pi is approximately: {pi:.5f}")  # limiting to 5 decimal places
print(f"The value of pi is approximately: {pi:.10f}")  # limiting to 10 decimal places
```
