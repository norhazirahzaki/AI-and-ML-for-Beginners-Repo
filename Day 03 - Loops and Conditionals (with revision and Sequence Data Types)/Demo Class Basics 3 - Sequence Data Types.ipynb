### Sequence Data Types in Python

A **sequence** is an ordered collection. You can:
- get length with `len(...)`
- access by **index** (first item is index `0`)
- take **slices** `seq[start:stop]`
- check membership with `"x" in seq`


Core sequence types you’ll use most:
1. **str** — text (immutable)
2. **list** — editable list of items (mutable)
3. **tuple** — fixed record (immutable)
4. **range** — sequence of numbers (efficient to store)
5. **dict** — collection of key-value pairs (mutable)
6. **set** — unordered collection of unique items (mutable)

```python
name = "Idrees"

print(name[0])
print(name[1])
print(name[2])
print(name[3])
print(name[4])
print(name[5])

# print(name[6]) # this will throw an error
```

## Tuple



- are used in places where you have fixed data
- where data should not change
- because tuples are immutable

```python
# unpacking
a,b,c,d = employee_id
print(a,b,c,d)
```

```python
employee_id = (101,202,303,404)
# empty tuple is 40 bytes in size
# first integer = 28
# two integer = 28+28 = 56


import sys
print(sys.getsizeof(employee_id))
```

```python
employee_id[2] = 5

# you cannot change the value of a tuple member or element
```

## Dictionary


- we save data using, key:value pairs
- 

```python
usernames = ("idrees","samson","ameer","yaya","idrees") 
```

```python
employees_ids = {
    usernames[0]:101,
    usernames[1]:202,
    usernames[2]:303,
    usernames[3]:404
}

# 64 bytes to an empty dictionary
# 224 bytes to the rest of the element of the dictionary
import sys
sys.getsizeof(employees_ids)
```

```python
employees_ids[usernames[1]] = 505
employees_ids
```

```python
# For loops on dictionaries

for key,values in employees_ids.items():
    print(key,"-->", values)

for key in employees_ids.keys():
    print(key,"-->", values)

for key in employees_ids.values():
    print(key,"-->", values)
```
