In normal programming, we often think in terms of:

- variables  x, y, z = 10, 20,30
- functions  def function(): 

In **OOP**, we think in terms of:

- **classes** ‚Üí blueprints / designs  
- **objects** ‚Üí actual things created from those blueprints 

Real-life analogy

- **Class**: `Student` (a general design ‚Äì name, age, grade, actions)  
- **Object**: Riya, Amit, Sneha (real students with actual data)

In Python:

- A **class** groups data (attributes) and functions (methods) together.
- An **object** is created from a class and has its own data.

Think about real life:

- You are a **person**
- Your car is an **vehicle**
- Your phone is a **device**

Each has:

- **Properties** (kind, color, age, brand, price)
- **Actions** (a car can honk, a phone can ring, computer plays music)

In OOP:

- A **class** is like a *blueprint* (e.g. the idea of a "Possesions")
- An **object** is a *real thing* made from that blueprint (e.g. Phone, Computer, Car)

We use classes and objects to organize code in a way that looks like how we think about the real world.


> Let's understand the following things:

1. Creating a **class** and making **objects**
2. **Encapsulation** (hiding internal details)
3. **Inheritance** (reusing and extending a class)
4. **Polymorphism** (same method, different behavior)

#### Example: Simple `Student` Class and Object

````python
# A very simple Student class

class Possesions:
    # __init__ is the constructor: it runs when we create a new Student
    # __init__ is a special method that runs automatically when an object is created
    def __init__(self, kind, age, color, brand): # first parameter, refers to the object
        # self refers to "this object"
        self.kind = kind    # attribute
        self.age = age      # attribute
        self.color = color  # attribute
        self.brand = brand  # attribute

    # a method (function inside a class)
    def details(self):
        print(f"I have a {self.kind}, It is {self.age} years old now. It is of color {self.color}. The brand is {self.brand}.")
````

````python
# Creating objects (instances) of Student
phone = Possesions("Phone", 16, "red","iphone")
car = Possesions("Car", 15, "red", "Volkwagon")

car.brand
car.color
````

````python
# Calling methods on objects
phone.details()
car.details()
````

#### Another Example



Think of **Dog** as a general class:

- Each dog has a **name**
- Each dog has an **age**
- Each dog can **bark**
- Each dog has a **birthday**



````python
# A simple class for a Dog

class Dog:
    # __init__ runs when we create a new Dog
    def __init__(self, name, age):
        # 'self' refers to the current object
        self.name = name    # property
        self.age = age      # property

    # A method (action) of the Dog
    def bark(self):
        print(f"{self.name} says: Woof! üê∂")

    def birthday(self):
        # Increase age by 1
        self.age += 1
        print(f"It's {self.name}'s birthday! Now {self.age} years old üéâ")


# Creating objects (dog1 and dog2 are objects of class Dog)
dog1 = Dog("Tommy", 3)
dog2 = Dog("Luna", 1)

# Using their methods
dog1.bark()
dog2.bark()

dog1.birthday()
dog2.birthday()
````

- `class Dog:` defines a **blueprint**
- `__init__` sets up properties when we create a new Dog
- `self.name` and `self.age` are **attributes** (data)
- `bark()` and `birthday()` are **methods** (actions)
- `dog1 = Dog("Tommy", 3)` creates an **object**

    <h4>2. Encapsulation
</h4> </div>

**Encapsulation** means:

- Keeping the **data** (variables) and the **methods** (functions) that work on that data **together** in a class.
- Hiding the ‚Äúinner details‚Äù from the outside and only exposing what we *want*.

In real life:

- You use a **phone**.
- You press buttons / touch the screen.
- You don't see the complex circuits inside.

In code:

- The **class** is like the phone.
- It has internal data that **should not be changed directly** from outside.
- We use **methods** to safely read or update that data.

In Python, we use a common convention:

- Names starting with `_` (single underscore), like `_age`, mean:
  > "This is internal. Please don't touch it directly from outside the class."


````python
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        # Age is "internal" by convention (starts with _)
        # We want other code to use our methods instead of touching _age directly
        # We still store it with self, but the _ tells other programmers:
        # "Don't change this directly from outside if you can avoid it."
        self._age = age

    def introduce(self):
        print(f"Hi! I'm {self.name}, a {self.species}, and I'm {self._age} years old. üêæ")

    # Getter method: safe way to read age
    def get_age(self):
        return self._age
    
    # setter method
    def set_age(self,new_age):
        self._age= new_age
        return f"The new age is {new_age}"

    # Method to have a birthday ‚Üí controlled way to change age
    def have_birthday(self):
        self._age += 1  # only this method decides HOW age changes
        print(f"It's {self.name}'s birthday! Now {self._age} years old üéâ")


# Create an animal
dog1 = Animal("Bruno", "Dog", 3)

dog1._age 
dog1.introduce()
print("Current age (via getter):", dog1.get_age())

dog1.have_birthday()
print("Age after birthday:", dog1.get_age())

# Technically, this still works, but it's NOT good style:
# Encapsulation is the IDEA that data should be changed via methods
dog1._age = 100  # breaks encapsulation idea
print("After changing _age directly:", dog1.get_age())
````

````python
dog1.get_age()
dog1.set_age(50)

dog1.get_age()
````

    <h4>3. Inheritance
</h4> </div>

**Inheritance** means:

- A new class (child) can reuse code from an existing class (parent)
- Child class can *add* new things or *change* some things

Now imagine:

- We have a general `Animal` class.
- We want more specific types:
  - `Dog`
  - `Cat`
  - `Bird`

They are all **Animals**, so they should:

- Reuse the `Animal` data (name, species, age)
- Add their own behaviours (like different sounds)

**Inheritance** lets a new class use an existing class as a base.

- `class Dog(Animal):` means  
  `"Dog is a kind of Animal"`

- We have a general **Animal** class
- Dogs, Cats, and Birds are all **animals**, but each makes a different sound

We will:

- Use `super().__init__()` to call the parent (`Animal`) constructor.
- Add a new method like `speak()` to each child.

````python
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self._age = age

    def introduce(self):
        print(f"Hi! I'm {self.name}, a {self.species}, and I'm {self._age} years old. üêæ")


# Dog is a specific kind of Animal
class Dog(Animal):
    def __init__(self, name, age):
        # Call the parent class __init__ to set up name, species, and age
        # 'super()' refers to the parent class (Animal)
        # we could use this as well: Animal.__init__(self, name, "Dog", age)
        super().__init__(name, "Dog", age)

    def speak(self):
        print(f"{self.name} says: Woof! üê∂")

# ------

# Cat is also a specific kind of Animal
class Cat(Animal):
    def __init__(self, name, age):
        super().__init__(name, "Cat", age)

    def speak(self):
        print(f"{self.name} says: Meow! üê±")


dog = Dog("Rocky", 4)
cat = Cat("Luna", 2)

dog.introduce()   # from Animal
dog.speak()       # from Dog

cat.introduce()   # from Animal
cat.speak()       # from Cat
````

- can we have the same name for two functions

- generally, create same name functions, whichever is the last one, is the final one

- if we keep same function names in different classes, both will work differently - 
> polymorphism

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h4>4. Polymorphism
</h4> </div>

**Polymorphism** means:

> The same method name can work in different ways for different classes.

> "Many forms"

In our example:

- Both `Dog` and `Cat`:
  - Are `Animal`s
  - Have a `speak()` method
- But:
  - Dog's `speak()` says **"Woof"**
  - Cat's `speak()` says **"Meow"**

So:

- The **method name** is the same: `speak()`
- The **behavior** is different, depending on the object‚Äôs class
- `Dog.speak()` says "Woof", and `Cat.speak()` says "Meow"

This is very powerful, because we can write code that works on **Animals** in general, without caring if it's a Dog or Cat.

````python
class Dog(Animal): # child class
    def speak(self): # same name method
        print(f"{self.name} says: Woof! üê∂")

class Cat(Animal): # child class
    def speak(self):
        print(f"{self.name} says: Meow! üê±")

class Cow(Animal): # child class
    def speak(self):
        print(f"{self.name} says: Moo! üêÑ")

# Create a list of different animals
animals = [
    Dog("Bruno", "dog",13),
    Cat("Kitty","cat",5),
    Cow("Gauri","cow",8),
]

# All have .speak(), but each behaves differently
for animal in animals:
    animal.speak()
````

### Tiny Guessing Game 

We will build a **very small game**:

- The computer thinks of a number between 1 and 5
- You have to guess the number
- The game will tell you:
  - "Too low"
  - "Too high"
  - or "You got it üéâ"
- We will wrap this in a **Game class**

````python
import random  # to let the computer pick a random number

class GuessGame:
    def __init__(self, low=1, high=5):
        # Save the range
        self.low = low
        self.high = high
        # Computer chooses a secret number once
        self.secret = random.randint(self.low, self.high)

    def play(self):
        print(f"I'm thinking of a number between {self.low} and {self.high}.")
        print("Can you guess it? ü§î")

        while True:
            # Ask the player for a guess
            guess = int(input("Your guess: "))

            # Compare guess with secret
            if guess < self.secret:
                print("Too low! üìâ Try again.")
            elif guess > self.secret:
                print("Too high! üìà Try again.")
            else:
                print("You got it! üéâ")
                break  # exit the loop when guessed correctly


# Create a game object and start the game
game = GuessGame()  # creates a game with numbers 1 to 5
game.play()
````
