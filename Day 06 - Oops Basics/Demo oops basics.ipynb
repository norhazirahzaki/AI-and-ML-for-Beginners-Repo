````python
# CamelCase - Pascal Notation - conventional ways to name class
# blueprint for object
# you define the rules for construction of an object in a class

class PossesionsIdrees:    # things in this class # type of object # age # color # brand
    # combination of different types of data and functions 
    # __init__ constructor - makes objects
    def __init__(self,kind, age, color, brand): # __init__ class, 
        self.kind = kind
        self.age = age
        self.color = color
        self.brand = brand

    # methods are functions wrapped inside classes

    def kind_of_possesion(self):
        print("The possesion is: ",self.kind)

    def age_of_possesion(self):
        print("The possesion is: ",self.kind)

    def history(self):
        x = f"I have a {self.kind}, {self.brand} and i have had it for {self.age}"
        return x
````

````python
# Phone Object
ph = PossesionsIdrees("Phone","4 years","Purple","iPhone 14pm")


# kind of doing the same thing? as functions
# but. Functions return values or data types
# but Classes return objects!!!!

x = ph.color # object attribute
print(x)

y = ph.history() # object method  
print(y)
````

# Oops concepts

1. encapsulation - box things up in a class, so everyone can not see them
2. inheritance - child inherits the parent class methods and properties
3. polymorphism - many forms - same methods, giving different outputs

````python
class Animal:
    def __init__(self,name, species, age):
        self.name = name
        self.species = species
        self._age = age # encapsulation rule - this can be changed - for changing i will create methods

    # methods
    def introduce(self):
        x = f"Hi, i am a {self.species} named {self.name}. I am {self._age} years old"
        print(x)
        return x
    
    # getter method
    def get_age(self):
        return self._age
    
    # setter method
    def set_age(self,new_age):
        self._age= new_age 
````

````python
animal1 = Animal("Tommy", "dog", 9)

animal1._age
````

````python
# let's inherit the properties on the parent Animal class
class Dog(Animal):         # dog will inherit properties of Animal
    def __init__(self,name, age):
        super().__init__(name, "Dog", age) 
        #Animal.__init__(name,,age)

    # methods?

class Cat(Animal):         # cat refers to the parent just like dog
    def __init__(self,name, age):
        super().__init__(name, "Cat", age)
````

````python
dog1 = Dog("Tommy",3)


cat1 = Cat("Kitty",3)
````
