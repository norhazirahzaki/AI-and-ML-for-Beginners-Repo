```python
# how do we handle files in python
# open, read, write, close

#"w" write
#"r" read
#"a" append

file = open("notes.txt","w") # create a file

# write something into file
file.write("This is my first line\n and i will not be defeated")
```

```python
file = open("notes.txt","r") # read a file
content = file.read()
print(content)
file.close
```

```python
# Better way: with (context manager)
with open("notes.txt", "r") as f:
    content = f.read()
    print(content)
# file is automatically closed here
```

```python
# reading a file line by line
with open("notes.txt", "r") as file:
    for line in file:
        print(line.strip())  # strip() removes spaces
```

```python
# Count how many lines are in a file

line_count = 0 

with open("notes.txt", "r") as f:
    for line in f:
        line_count += 1

print("Total number of lines:", line_count)
```

#### problem

    repeat until a user enters 'q'
    Ask users to enter 5 numbers
    user enters a number
    add the number to a list
    display numbers from the list divisible by 5
    and ask user to enter 5 numbers again.

```python


numbers = []

while True:       # infinite loop
    choice = input("Press 'q' to quit or enter 5 numbers: ") # will be a string

    if choice.lower() == 'q':    # taking Q as well
        print("Exiting the program. Byeeeee!")
        break

    # Ask the user to enter 5 numbers
    print("Please enter 5 numbers:")
    
    for i in range(5):
        num = input(f"Enter the number {i + 1}: ")
        numbers.append(int(num))  # add the number to the list

    print("\nNumbers in the list that are divisible by 5:")
    
    found = False
    for n in numbers:
        if n % 5 == 0:
            print(n)
            found = True

    if not found:
        print("No numbers divisible by 5 in the list.")

    print("_" * 30)  # just a separator for readability
```
