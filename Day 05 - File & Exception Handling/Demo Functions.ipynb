#### Functions that perform a task

- we can encapsulate, group together many lines of code
- function can do a task
- or return a value
- save the returned value in a variable

```python
def greet():       # function definition
    print("Hello")
```

```python
def greet(name): # parameters
    print("Hello ",name)


# function call
greet("idrees") # arguments
```

```python
def greet(name="idrees"): # default parameter
    
    #message = "Hello"
    
    return f"Hello {name}"


# function call
s = greet("sam") # arguments

y = greet("yaya")

i = greet()

print(s)
print(y)
print(i)
```

```python
# create a function that calculates returns something

def sum(a,b):
    print(f"a is: {a} and b is: {b}")
    return a+b

a = sum(5,9)

print(a)
```

```python
# keyword arguments

def sum(a=0, b=0, c=0 ):
    print(a)
    print(b)
    print(c)
    print("--------")
    return a+b+c

#a = sum(b=9, c = 4, a=5, ) # keyword arguments
#print(a)
```

```python
sum(3,4,5) # the number of arguments

```

#### Lambda functions
- anonymous - no name
- one line

```python
# create a function that calculates returns something
def sum(a,b):
    return a+b

a = sum(5,9)
print(a)

# lambda parameters : expression 

b = lambda a,b : a + b 
print(b(2,5))
```

#### Recursion

- when a function calls its own self again

- 2 key parts:
    - base case - exit condition
    - expression, or what you want to do

5,4,3,2,1 

```python
def countdown( a ):                # a = 10        a =10 -1 = 9   a = 8
    # base condition
    if a == 0:
        print("Let's celebrate!!")
        return a 
    # solve a problem
    print("T minus: ",a)          # prints 10     print 9
    countdown( a - 1)             # a = 10        a = 9    
    
```

```python
def countdown( a ):              
    if a == 0:
        return a 
    print("T minus: ",a)     
    countdown( a - 1) 

countdown(10)
```
