</h3> </div>

###  What is a “Relationship” Between Two Variables?

In data, we often want to know:

- If one value **increases**, does the other also increase?
- Or, if one value **increases**, does the other **decrease**?
- Or, are they completely **unrelated**?

Examples in real life:

- More **hours studied** → higher **exam marks** (usually positive relationship)  
- More **junk food** → worse **health** (often negative relationship)  
- Shoe size and IQ → probably no useful relationship

To measure this “moving together” idea, we use:

1. **Covariance**
2. **Correlation**

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Covariance
</h4> </div>

**Covariance** tells us how two variables move together.

- **Positive covariance**  
  When one goes up, the other usually goes up.  
  When one goes down, the other usually goes down.

- **Negative covariance**  
  When one goes up, the other usually goes down.

- **Covariance near 0**  
  No clear pattern. They don’t move together in any clear way.


> Problem:  
Covariance depends on the **scale** of data, so the numbers are hard to compare between different pairs of variables.

That’s why we prefer **correlation** in practice.

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Correlation 
</h4> </div>

**Correlation** improves on covariance by:

- Scaling the value to always be between **-1 and 1**
- So it is easier to interpret and compare

Interpretation:

- **+1** → perfect positive linear relationship  
- **-1** → perfect negative linear relationship  
- **0** → no linear relationship

Examples:

- Correlation ≈ **+0.9**  
  Strong positive relationship (e.g., hours studied vs marks)

- Correlation ≈ **-0.8**  
  Strong negative relationship (e.g., days absent vs marks)

- Correlation ≈ **0.1**  
  Very weak or no clear linear relationship

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Tiny Example  
</h4> </div>

Imagine we have 3 students:

| Student | Hours Studied | Marks |
|--------|---------------|-------|
| A      | 2             | 50    |
| B      | 4             | 70    |
| C      | 6             | 90    |

As **hours studied** increase, **marks** increase.
- Covariance will be **positive**
- Correlation will also be **positive**, probably close to +1

If instead the marks were:

| Student | Hours Studied | Marks |
|--------|---------------|-------|
| A      | 2             | 90    |
| B      | 4             | 70    |
| C      | 6             | 50    |

Now, as hours studied increase, marks **decrease**.
- Covariance will be **negative**
- Correlation will be **negative**, maybe close to -1

We won’t compute the formulas by hand.  
We will let Python do the calculation and focus on **reading and understanding** the results.

> What is in the `tips` dataset?

- `tip` – tip given (in dollars)

- `sex`, `smoker`, `day`, `time` – other info (categorical)

```python
# Check basic info about the dataset

tips.info()
```

```python
# Summary statistics of numeric columns

tips.describe().round(2)
```

### Covariance in Pandas

We can use `.cov()` on a DataFrame to compute the **covariance matrix**.

- Each cell shows the covariance between two columns.
- We will look at: `total_bill`, `tip`, `size`.

```python
# Select only the numeric columns we care about

num_cols = ["total_bill", "tip", "size"]
tips_num = tips[num_cols]

tips_num.head()
```

```python
# Covariance matrix

cov_matrix = tips_num.cov()
cov_matrix.round(2)
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Interpreting the Covariance Matrix (Simple)

Look at the values:

- Covariance between `total_bill` and `tip`:
  - Is it positive or negative?
  - What does that tell us?

- Covariance between `size` and `total_bill`:
  - Is it positive or negative?

We only care about the **sign** here:
- Positive → they move in the same direction
- Negative → one goes up, the other tends to go down

### Correlation in Pandas

Now we use `.corr()` to compute the **correlation matrix**.

This is usually more useful because:
- Values always between **-1 and 1**
- Easier to compare


```python
# 7.1 Correlation matrix

corr_matrix = tips_num.corr()
corr_matrix.round(2)
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Interpreting the Correlation Matrix

- How strongly is `total_bill` correlated with `tip`?
  - Is the value closer to 0, 0.5, or 1?

- How strongly is `size` correlated with `total_bill`?
- Are there any pairs with very weak correlation (near 0)?

Remember:

- Closer to **+1** → strong positive relationship  
- Closer to **-1** → strong negative relationship  
- Around **0** → weak or no clear linear relationship

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3> Visualizing Relationships: Scatter Plots and Heatmaps
    </h3> </div>

#### Scatter Plots

Scatter plots help us **see** correlation:

- If dots go up together → positive correlation
- If dots go in opposite directions → negative correlation
- If dots form a cloud → little or no correlation

```python
# Scatter plot: total_bill vs tip

plt.figure()
sns.scatterplot(data=tips, x="total_bill", y="tip")

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

- Does `total_bill` vs `tip` look like a **positive** relationship?
- Does `size` vs `total_bill` make sense? (More people → bigger bill?)
- Are the plots consistent with the **correlation values** we saw?

#### Correlation Heatmap

A **heatmap** is a colored grid that shows all correlations in one picture.

- Darker/stronger colors → stronger correlation
- Color sign (e.g., red vs blue) → positive or negative

```python
# Plot correlation heatmap

plt.figure(figsize=(6, 4))
sns.heatmap(
    corr_matrix,
    annot=True,      # show numbers
    fmt=".2f",       # 2 decimal places
    cmap="coolwarm", # color map
    vmin=-1, vmax=1  # fixed scale from -1 to 1
)
plt.title("Correlation Heatmap – tips dataset")
plt.show()
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Reading the Heatmap

- Which pair has the **highest positive** correlation?
- Is there any **negative** correlation here?
- Are there any near **zero**?

> "As total_bill increases, tip also increases, because their correlation is about 0.xx (positive)."
