```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
tips_numerical = df[['total_bill','tip','size']]
tips_numerical.head()
```

```python
catagorical = df.select_dtypes(exclude= 'number')
catagorical.head()
```

```python
numerical = df.select_dtypes(include='number')
numerical.head()
```

```python
covariance_tips = tips_numerical.cov()

covariance_tips.round(2)
```

```python
sns.scatterplot(data=tips_numerical, x ='total_bill', y = 'tip' )
```

```python
sns.pairplot(data=df, hue= 'sex', corner=True)
```

#### corelation matrix

.cov - covariance matrix 

```python
# # pearsons corelation

covariance_tips = tips_numerical.corr(method='pearson')

covariance_tips.round(2)
```

```python
# Heat map to visualise the matyrix 

sns.heatmap(covariance_tips, 
            annot= True,
            cmap= "Greens", # coolwarm inferno crest plasma Purples Greens
            vmin= 0, vmax= 1
            )
```
