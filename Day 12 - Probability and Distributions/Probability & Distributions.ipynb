    <h3> What is Probability?

We use **probability** to measure how likely something is to happen.

- A **random experiment** is an action with an uncertain result.  
  Examples:
  - Tossing a coin  
  - Rolling a dice  
  - Picking a card from a deck  

- An **outcome** is one possible result of the experiment.  
  - Coin: `Head` or `Tail`  
  - Dice: `1, 2, 3, 4, 5, 6`

- An **event** is a collection of outcomes.  
  - “Getting an even number” when rolling a dice → {2, 4, 6}  
  - “Getting a Head” when tossing a coin → {H}

If all outcomes are equally likely:

> {Probability of an event} =  {Number of favorable outcomes} / {Total number of possible outcomes}


- Probability of getting a Head in 1 fair coin toss:
    - P(Head) = 1 / 2

- Probability of getting a 4 on a fair dice:
    - P(4) = 1 / 6

- Probability of getting an even number (2,4,6) on a fair dice:
    - P(even) = 3 / 6 = 1 / 2

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> What is a Probability Distribution?
    

A **probability distribution** tells us:

> For each possible value of a random variable, what is the probability of that value?

- A **random variable** is a variable whose value comes from a random experiment.
  - Example:  
    - X = number you get when you roll a dice  
    - Y = number of Heads in 3 coin tosses

A **probability distribution** can be:

- **Discrete** – the variable takes specific, separate values  
  (e.g., 0, 1, 2, 3 heads in 3 coin tosses)

- **Continuous** – the variable can take any value in a range  
  (e.g., height, weight, marks out of 100, time taken to finish a task)

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3> Binomial Distribution 

The **Binomial distribution** describes the number of successes in a fixed number of **independent** yes/no trials.

We use it when:

1. There are only **two outcomes** each time (often called “success” and “failure”).  
   - Example: Head/Tail, Yes/No, Pass/Fail

2. The probability of success (**p**) is the same each time.  
   - Example: For a fair coin, p = 0.5

3. The trials are **independent** (what happens in one trial does not affect others).

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  Example :

You toss a fair coin **3 times**.  
Let X = number of Heads.

Possible values of X: 0, 1, 2, 3.

We can list all possible sequences:

- T T T → 0 Heads  
- H T T → 1 Head  
- T H T → 1 Head  
- T T H → 1 Head  
- H H T → 2 Heads  
- H T H → 2 Heads  
- T H H → 2 Heads  
- H H H → 3 Heads  

So we can count:

- P(X = 0) = 1/8  
- P(X = 1) = 3/8  
- P(X = 2) = 3/8  
- P(X = 3) = 1/8  

This is an example of a **Binomial distribution** with:
- Number of trials `n = 3`
- Probability of success (Head) `p = 0.5`

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3> Normal Distribution

The **Normal distribution** is a continuous distribution with a **bell-shaped curve**.

Many real-world measurements are **approximately** normal:

- Heights of people  
- Exam marks (if enough students)  
- Measurement errors  
- Blood pressure, IQ scores, etc.

Key ideas:

- The curve is **symmetric** around the mean.
- Most values cluster around the **mean**.
- Fewer values are found far away from the mean.
- It is defined by two parameters:
  - **Mean (μ)** – center of the distribution  
  - **Standard deviation (σ)** – spread (how wide the bell is)

We will not dive into formulas today.  
We will focus on **seeing** and **simulating** the Normal distribution.

    <h4> Let’s Use Python to Simulate and Visualize

We will:

1. Simulate coin tosses (basic probability idea)
2. Simulate a **Binomial distribution**:
   - Number of Heads in multiple coin tosses
3. Simulate a **Normal distribution**:
   - Example: exam marks around 70 with some spread
4. Plot histograms to see the shapes

```python
# Import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nicer
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (7, 5)

# Set a random seed so results are reproducible
np.random.seed(42)
```

#### Simple Simulation: Coin Toss

We’ll start with something simple:

- Toss a fair coin many times
- Estimate the probability of getting Heads and Tails

```python
# Simulate coin tosses

n_tosses = 1000

# We'll use 1 for Head, 0 for Tail
coin_tosses = np.random.choice([0, 1], size=n_tosses, p=[0.5, 0.5])
# np.random.choice([0, 1], ...) - Randomly choose values from the list [0, 1]
# Do this n_tosses times and give me an array of that many values.
# p=[0.5, 0.5] - Use 50% chance for 0 and 50% chance for 1.


# Count Heads and Tails
unique, counts = np.unique(coin_tosses, return_counts=True)

# turn them into a dictionary
# zip combines the two lists “side by side”
results = dict(zip(unique, counts))

print(results)
```

```python
# Convert results to probabilities

prob_head = results.get(1, 0) / n_tosses
prob_tail = results.get(0, 0) / n_tosses

# dict.get(key, default_value)
# if key exists, return its values, otherwise return the default value

print(f"Estimated P(Head) ≈ {prob_head:.3f}")
print(f"Estimated P(Tail) ≈ {prob_tail:.3f}")
```

```python
# Visualize Heads vs Tails

labels = ["Tail (0)", "Head (1)"]
values = [results.get(0, 0), results.get(1, 0)]

plt.bar(labels, values)
plt.title("Number of Heads and Tails in Coin Tosses")
plt.ylabel("Count")
plt.show()
```

#### Observation

- Are the counts of Heads and Tails close to each other?
- As `n_tosses` becomes very large, the estimated probabilities get closer to 0.5.

This is an example of **experimental probability** approaching **theoretical probability**.

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Simulating a Binomial Distribution

Now let’s use a built-in function to simulate a **Binomial** random variable.

We will model:

> X = number of Heads in 10 coin tosses  
> (n = 10 trials, p = 0.5)

We will repeat this experiment many times (e.g., 2000 times) and see how many times we get 0, 1, 2, ..., 10 Heads.


```python
# Simulate Binomial: number of Heads in 10 tosses

n_trials = 10   # number of coin tosses per experiment
p = 0.5         # probability of Head in each toss
n_experiments = 2000  # how many times we repeat the experiment

# Simulate
binomial_samples = np.random.binomial(n=n_trials, p=p, size=n_experiments)

binomial_samples[:20]  # show first 20 results
```

```python
values
```

```python
# Count how often each result (0,1,...,10 Heads) occurs

values, counts = np.unique(binomial_samples, return_counts=True)

for v, c in zip(values, counts):
    print(f"Number of Heads = {v}: occurred {c} times")
```

```python
# Plot histogram of the Binomial distribution

plt.bar(values, counts)
plt.xlabel("Number of Heads in 10 tosses")
plt.ylabel("Frequency (out of 2000 experiments)")
plt.title("Binomial Distribution: n=10, p=0.5")
plt.xticks(range(0, n_trials + 1))
plt.show()
```

#### Observation

- Most of the time, we get around **5 Heads** (near n * p = 10 * 0.5 = 5).
- Very few experiments have 0 Heads or 10 Heads.
- This is the **shape** of a Binomial distribution for n = 10, p = 0.5.

    <h4> Simulating a Normal Distribution

Now let’s simulate data that *approximately* follows a **Normal distribution**.

Example:

> Imagine exam marks for 1000 students:  
> - Mean mark ≈ 70  
> - Standard deviation ≈ 10  

We’ll generate random marks and see the bell-shaped curve.

```python
# Simulate normally distributed exam marks

n_students = 1000
mean_marks = 70     # center
std_marks = 10      # spread

marks = np.random.normal(loc=mean_marks, scale=std_marks, size=n_students)

# Clip marks to be within 0 and 100
marks = np.clip(marks, 0, 100)

marks[:20]  # show first 20 marks
```

```python
# Basic stats of the simulated marks

print("Simulated mean:", np.mean(marks).round(2))
print("Simulated std dev:", np.std(marks).round(2))
```

```python
# Plot histogram of marks

sns.histplot(marks, bins=20, kde=True)
plt.title("Simulated Exam Marks (Approx Normal Distribution)")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

#### Observation

- The histogram looks roughly like a **bell curve**.
- Most students are around the mean (e.g., 60–80).
- Very few students are very low or very high.

This is the typical shape of a **Normal distribution**.

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Comparing Binomial and Normal

- **Binomial distribution**:
  - Discrete (0, 1, 2, ..., n)
  - Used for counts of successes in fixed number of yes/no trials
  - Example: number of Heads in 10 tosses

- **Normal distribution**:
  - Continuous (any value in a range)
  - Used for many real-valued measurements
  - Example: heights, marks, weights

As n becomes large and p is not too close to 0 or 1,  
a **Binomial distribution** can start to look similar to a Normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Settings ---
mean_height = 165   # cm
std_height = 7      # cm
n_people = 10000    # how many heights to simulate

# Generate random heights
heights = np.random.normal(loc=mean_height, scale=std_height, size=n_people)

# Check the sample mean and std
print("Sample mean:", heights.mean())
print("Sample std dev:", heights.std())

# Plot the histogram
plt.hist(heights, bins=30)
plt.xlabel("Height (cm)")
plt.ylabel("Number of people")
plt.title("Simulated Heights: mean=165, std=7")
plt.show()
```
