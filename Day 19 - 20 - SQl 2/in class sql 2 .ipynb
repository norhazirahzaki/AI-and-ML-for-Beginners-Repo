```python
select * from x
select * from y

select idx,name
from x

select idy, address
from y

-- inner join --
SELECT x.idx, y.nickname
FROM y
INNER JOIN x ON x.idx = y.idy

-- right join
SELECT x.idx, y.nickname
FROM x
RIGHT JOIN y ON x.idx = y.idy

-- left join
SELECT x.idx, y.nickname
FROM x
LEFT JOIN y ON x.idx = y.idy

-- full outer join
SELECT x.idx, y.idy, y.nickname
FROM x
FULL OUTER JOIN y ON x.idx = y.idy



```
