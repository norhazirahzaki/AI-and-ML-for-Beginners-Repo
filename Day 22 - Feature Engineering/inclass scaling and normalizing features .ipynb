```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Raw data: Age vs. Yearly Income
data = pd.DataFrame({
    'age': [0, 50, 100],
    'income': [20000, 50000, 100000]
})
data
```

```python
scaler = MinMaxScaler()

normalized_data = scaler.fit_transform(data)
pd.DataFrame(normalized_data)
```

- Minimum value =  0
- Max value = 1

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()

standardized_data = std_scaler.fit_transform(data)
pd.DataFrame(standardized_data)
```

- Minimum value = -1.111168
- Max value = 1.313198

- whole data has a mean of 0
- variance of 1
