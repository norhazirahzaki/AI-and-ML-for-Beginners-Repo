```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# One hot encoding for species

penguinsohe = penguins.copy()


penguins_encoded = pd.get_dummies(
    penguinsohe,
    columns = ['species'],
    drop_first = True
    # drops the first catagory column
    # only one column will be enough to know about both the catagories
)

# kind of redundant to keep all the catagory columns

penguins_encoded.sample(5)
```

```python
# sklearn onehotencoder

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)

penguins_oh_encoded = ohe.fit_transform(penguinsohe[['species']])
print(penguins_oh_encoded)

df= pd.DataFrame(penguins_oh_encoded)
df


penguinsohe[['adelie_ohe','chinstrap_ohe','gentoo_ohe']] = penguins_oh_encoded
penguinsohe.sample(5)
```
