```python
# to plot a line in python
import matplotlib.pyplot as plt

```

```python
x = np.linspace(-100,100,100)

# y = m x + c
# y = 2 * x - 25

y = x ** 2
y1 = 10 * x ** 2
y2 = ( 10 * x ** 2 ) + 10000 

plt.plot(x, y)

plt.plot(x, y1, 'blue')

plt.plot(x, y2, 'green')

# plt.xlim(-200, 200)
# plt.ylim(-1000,10000)
plt.axhline(0, color='red', linewidth = 0.5)
plt.axvline(0,color='red', linewidth = 0.5)
# plt.grid()
```

```python
x = np.linspace(-100,100,100)

y = x ** 2
y_left  = (x + 50) ** 2 # shifting to left
y_right = (x - 50) ** 2 # shifting to right
y_up    = (x) ** 2  + 1000 # shifting to up
y_down  = (x) ** 2  - 1000 # shifting to down

plt.plot(x, y, linestyle="--")
plt.plot(x, y_left, color= 'green')
plt.plot(x, y_right, color= 'purple')
plt.plot(x, y_up, color= 'yellow')
plt.plot(x, y_down, color= 'brown')

plt.xlim(-200, 200)
plt.ylim(-1000,10000)
plt.axhline(0, color='red', linewidth = 0.5, linestyle="--")
plt.axvline(0,color='red', linewidth = 0.5, linestyle="--")
# plt.grid()
```

```python
x = np.linspace(0,10,100)

y = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y, color= 'purple', label = 'sin')
plt.plot(x, y2, color= 'green', label = 'cos')

plt.legend()
plt.axhline(0, color='red', linewidth = 1)
plt.axvline(0,color='red', linewidth = 1)
plt.grid()
plt.legend()
```
