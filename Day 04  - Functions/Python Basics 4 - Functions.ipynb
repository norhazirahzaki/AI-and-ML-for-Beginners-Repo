#### What is a Function?
A function is a reusable block of code that performs a specific task. Functions help organize code, make it easier to read, and allow you to reuse code without rewriting it.

A function is like a small machine or a recipe.

- You:

- Give inputs (called parameters or arguments),

- The function does some work,




### Defining a Function
You define a function using the `def` keyword, followed by the function name and parentheses. Inside the parentheses, you can specify **parameters** (inputs) for the function.

#### *args - Variable number of arguments

Sometimes we wish to have functions that can take as many inputs as needed for a situation
- eg. 

````python
def multiply(x,y):
    return x*y

multiply(3,2) # 6 - works

# but what about multiple parameters
#multiply(2,3,4,5) # error - won't work because more arguments that required

# let's try this:

def multiply(*numbers): # use a plural name to indicate many parameters
    print(numbers)

multiply(3,4,2,3,5)
# so multiply(numbers) becomes a tuple
# we get a tuple, which is an iterable!!
# we cannot modify this tuple tho
````

````python
# we can then modify the function to return an actual value too
# which we can store in variables

def multiply(*numbers):
    total = 1
    for number in numbers:
        #number = number * total
        total *= number

    return total

print(multiply(3,4,2))
````

````python
# ANother function with Variable Arguments

def sum_all(*numbers): # allows the function to accept any number of arguments as a tuple.
    
    total = 0  # Initialize a variable to hold the total sum
    for num in numbers:  # Iterate over each number in the arguments
        total += num  # Add each number to the total
    return total  # Return the total sum

# Call the function with multiple arguments
total_sum = sum_all(1, 2, 3, 45, 4, 5)
print(total_sum) 
````

### Recusrion - function calling itself 
- Recursion is when a function calls itself.

- You break a big problem into smaller versions of the same problem.

- But one must do it in a controlled way

- Two key parts:

    - Base case â†’ when to stop. 
        - When the problem is so small that you can answer it directly.
        - This stops the recursion.

    - Recursive case â†’ do a small part, then call the function again.

```python
def function():
    if (a very small case):
        return answer
    else: 
        return (function with a smaller input)

````python
# create a countdown function

def countdown(n):
    # base condition
    if n == 0:
        print("Done!!")
        return n
    # problem to solve, and recursion
    print(n)
    countdown(n-1)

countdown(6)

    
````

Example:

- To sum numbers from 1 to 5:

    - Take 5, then add the sum of 1 to 4.

    - To get sum of 1 to 4, take 4, then add sum of 1 to 3.

    - Keep going until you hit 1.

````python
# a function that adds previous numbers

def sum_prev(n):
    # base condition
    if n == 0:
        return n
    
    # problem to solve, and recursion
    return n + countdown(n-1)

sum_prev(10)
````

````python
# a function that returns the factorial of a number

def factorialof(n):
    # case to help stop
    if n == 0:
        return 1 # if you keep n,then all numbers are multipled by 0 in end
    
    # return the problem to solve
    return n * factorialof(n-1)

factorialof(5)
````

````python
# Simple Solution to exercise

def factorial(n):
    # This function calculates the factorial of a positive integer n
    result = 1  # Initialize the result variable to 1
    for i in range(1, n + 1):  # Loop from 1 to n (inclusive)
        result *= i  # Multiply result by the current number i
    return result  # Return the final factorial result

# Example usage
number = 5  # Set the number for which we want to calculate the factorial
print(f"The factorial of {number} is {factorial(number)}.")  # Output the result
````

````python
# Exact solution to exercise

def factorial(n):
    # This function calculates the factorial of a positive integer n
    if n < 0:
        return "Factorial is not defined for negative numbers"  # Check for negative input
    elif n == 0 or n == 1:
        return 1  # The factorial of 0 or 1 is 1
    else:
        result = 1  # Initialize a variable to hold the factorial result
        for i in range(2, n + 1):  # Loop from 2 to n (inclusive)
            result *= i  # Multiply result by the current number i
        return result  # Return the final factorial result

# Example usage
number = 5  # Set the number for which we want to calculate the factorial
factorial_result = factorial(number)  # Call the factorial function
print(f"The factorial of {number} is {factorial_result}.")  # Output the result
````

````python
# hint implementation

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        print("Error: Cannot divide by zero!")
        return None
    return a / b
````

````python
# actual code

def calculator():
    print("Welcome to the Python Calculator ðŸ§®")
    print("Operations: +  -  *  /")
    print("Type 'q' to quit.\n")

    while True:
        op = input("Choose operation (+, -, *, /) or q to quit: ").strip()

        if op == 'q':
            print("Goodbye! ðŸ‘‹")
            break

        if op not in ['+', '-', '*', '/']:
            print("Invalid operation. Try again.\n")
            continue

        # get numbers from user
        try:
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
        except ValueError:
            print("Please enter valid numbers!\n")
            continue

        # perform operation
        if op == '+':
            result = add(a, b)
        elif op == '-':
            result = subtract(a, b)
        elif op == '*':
            result = multiply(a, b)
        elif op == '/':
            result = divide(a, b)

        if result is not None:
            print("Result:", result, "\n")

# run calculator
calculator()
````

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h4>1. Imperative Programming
</h4> </div>

**Definition**: This paradigm focuses on commands that change a programâ€™s state. It involves writing sequences of statements that lead to a desired outcome.

**Characteristics**: 
  - Uses statements to change program state.
  - Follows a step-by-step approach.

````python
# Example
total = 0
for i in range(1, 6):
    total += i  # Add i to total
print(total)  # Output: 15
````

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h4>2. Procedural Programming
</h4> </div>

**Definition**: A subtype of imperative programming that structures code into procedures (or functions). It promotes code reuse and organization.

**Characteristics**: 
  - Uses functions to encapsulate behavior.
  - Supports modularity.

````python
#example 

def calculate_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

print(calculate_sum(5))  # Output: 15
````

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h4>3. Object-Oriented Programming (OOP)
</h4> </div>

**Definition**: This paradigm is based on the concept of "objects," which are instances of classes. OOP promotes encapsulation, inheritance, and polymorphism.

**Characteristics**: 
  - Encapsulation: Bundling data and methods that operate on the data within one unit.
  - Inheritance: Mechanism to create a new class using the properties of an existing class.
  - Polymorphism: Ability to use a single interface for different data types.

````python
# Example
class Dog:
    def __init__(self, name):
        self.name = name
    
    def bark(self):
        return f"{self.name} says woof!"

my_dog = Dog("Buddy")
print(my_dog.bark())  # Output: Buddy says woof!
````

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h4>4. Functional Programming
</h4> </div>

**Definition**: This paradigm treats computation as the evaluation of mathematical functions and avoids changing state or mutable data.

**Characteristics**: 
  - First-class functions: Functions can be passed as arguments, returned from other functions, and assigned to variables.
  - Higher-order functions: Functions that operate on other functions. 
  - Immutability: Data is immutable.

````python
# Example
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squares = list(map(square, numbers))  # Apply the square function to each element
print(squares)  # Output: [1, 4, 9, 16]
````

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h4>5. Declarative Programming
</h4> </div>

**Definition**: This paradigm expresses the logic of a computation without describing its control flow. It focuses on what the program should accomplish rather than how to achieve it.

**Characteristics**: 
  - Focuses on the outcome rather than the procedure.
  - Commonly used in database querying and configuration management.

**Example:**
  - SQL for querying databases is a common example, while in Python, you can use libraries like Pandas to filter data in a declarative manner.


````python
#Example
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Score': [85, 90, 95]}
df = pd.DataFrame(data)

# Filter where Score is greater than 90
high_scores = df[df['Score'] > 90]
print(high_scores)
````
