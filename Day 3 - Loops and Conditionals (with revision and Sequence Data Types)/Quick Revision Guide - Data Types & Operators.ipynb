#### 1) Core Data Types 
- **str**: text, immutable â†’ slicing, methods (`lower`, `split`, `join`)
- **list**: ordered, mutable â†’ `append`, `pop`, slice
- **tuple**: ordered, immutable â†’ safe fixed records
- **set**: unordered, unique elements â†’ math ops (`| & -`)
- **dict**: keyâ€“value mapping â†’ `get`, `keys`, `values`, `items`

```python
# ---- Strings (immutable)
s = "Python for Data"
print(s[:6], s[-4:])           # 'Python' 'Data'
print(s.lower(), s.split())    # lowercase and token list
print("-".join(["A", "B", "C"]))

# ---- Lists (mutable)
cities = ["Delhi", "Mumbai", "Kolkata", "Chennai"]
cities[2] = "Bengaluru"        # mutate - change the value
print(cities[:2])              # slicing - first 2 elements
cities.append("Pune")          # adds an element at the end

print("after append+pop:", cities, "| popped:", popped)

# ---- Tuples (immutable)

# person[1] = 26  # <-- would error (immutability)
print(person, type(person))

# ---- Sets (unique, unordered)
nums = [1, 2, 2, 3, 3, 3]

print("unique:", uniq)         # {1, 2, 3}
print({1,2,3} | {3,4}, {1,2,3} & {2,3,5}, {1,2,3} - {2})

# ---- Dicts (keyâ€“value)

    print(k, "â†’", v)

```

#### 5) Micro Project â€” â€œMini Cartâ€
- Given a `prices` dict and a `cart` dict (itemâ†’qty), compute:
  - **subtotal** (sum of priceÃ—qty)
  - **unique item count** (via set/cart keys)
  - Check if `"coupon"` is in cart (membership)
  - Use `+=` for accumulation

````python

# ---- 8) Micro Project â€” Mini Cart
prices = {"apple": 30, "banana": 10, "milk": 45, "bread": 40}
cart   = {"apple": 3, "milk": 1, "bread": 2}  # item -> qty

subtotal = 0
for item, qty in cart.items():
    subtotal += prices.get(item, 0) * qty

unique_items = len(set(cart.keys()))
has_coupon = "coupon" in cart

print(f"Subtotal: â‚¹{subtotal}")
print("Unique items:", unique_items)
print("Coupon in cart?", has_coupon)
````
