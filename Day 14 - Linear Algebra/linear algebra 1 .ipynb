```python
import numpy as np


arr1D = np.array([1,  2,  3,  4,  5])
```

```python
# arrays - attribute
arr1D.ndim  # what are the dimensions of the array

arr1D.shape  # Rows , no columns

```

- attribute - only accessing information - do not need to pass arguments
- method - that takes an input - ( ) are mustttttt

- Arrays are better to represent matrices than lists
    1. only same dimension rows are allowed like in matrices. Shape should be consistent. 
    2. Broadcasting 

```python
arr2D = np.array([[1,   2,   3,   4,    5],
                  [10,  20,  30,  40,  50] ])

arr2D.shape
```

```python
arr2D.ndim  # attribute


arr3D = np.array([

    [[1,   2,   3,   4,    5],[10,  20,  30,  40,  50]], # 1
    [[1,   2,   3,   4,    5],[10,  20,  30,  40,  50]], # 2
    [[1,   2,   3,   4,    5],[10,  20,  30,  40,  50]], # 3
    [[1,   2,   3,   4,    5],[10,  20,  30,  40,  50]]  # 4
    
    ])

arr3D.ndim

arr3D.shape
```
